```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rental_raw = pd.read_csv("Apartment Prices Granular.csv")
```

```python
rental_raw_2 = rental_raw.replace(0, pd.NA)
rental_raw_3 = rental_raw_2.dropna(axis=1, thresh=5)
rental_raw_3.drop(columns=['Champaign-Urbana-IL'], inplace=True)
df = rental_raw_3.copy()
# rental_raw_3
```

```python
std_dev = df.std()

# Filter columns where standard deviation is less than 100
selected_columns = std_dev[std_dev < 100].index

# Get the subset of the DataFrame with selected columns
df_filtered = df[selected_columns]
df_filtered

# Drop columns with low variation from the original DataFrame
df.drop(columns=selected_columns, inplace=True)
```

```python
std_dev = df.std()

# Filter columns where standard deviation is less than 100
selected_columns = std_dev[std_dev > 750].index

temp = df[selected_columns]
plt.figure(figsize=(8, 6))
plt.xlabel('Variance')
plt.ylabel('Frequency')
plt.title('Variance of locations')
plt.hist(temp.var())
temp
```

```python
# Filter columns where standard deviation is less than 100
selected_columns = std_dev[std_dev < 1000].index

# Get the subset of the DataFrame with selected columns
df_filtered = df[selected_columns]
df_filtered
```
